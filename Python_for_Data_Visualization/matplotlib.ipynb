```python
import matplotlib.pyplot as plt
```

```python
# this line helps us to skip - plt.show () method each time.
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x ** 2
```

```python
plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel('Y Axis')
plt.title('Title')
plt.show()
```

```python
# Subplot or Multiple plot Graph
plt.subplot(1,2,1) # represents no of row - 1 , column - 2 and graph no is 1
plt.plot(x,y,'r')

plt.subplot(1,2,2) # represents no of row - 1 , column - 2 and graph no is 2
plt.plot(x,y,'b')
```

```python
# Object oriented Method

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel("X Lable")
axes.set_ylabel('Y Lable')
axes.set_title('Title')
axes.plot(x,y,'r')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.6,0.2])

axes1.plot(x,y,'r')
axes1.set_title('Larger Plot')

axes2.plot(y,x,'b')
axes2.set_title('Small Plot')
```

```python
fig, axes = plt.subplots(1 ,2)

axes[0].plot(x,y,'r')
axes[0].set_title("First Plot")

axes[1].plot(x,y,'b')
axes[1].set_title("Second Plot")

# recover from overlapping Value
plt.tight_layout()
```

# Figure size (in inches ) and DPI (dots per inch)

```python
fig = plt.figure(figsize=(7,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig , axes = plt.subplots(nrows=3,ncols=1,figsize=(7,4))

axes[0].plot(x,y)
axes[1].plot(y,x)
axes[2].plot(x,x)

plt.tight_layout()
```

```python
# To save Graph or plot as Image
fig.savefig('my_graph.png')
```

# Legend Usage 

```python
fig = plt.figure()

axes_leg = fig.add_axes([0,0,1,1])

axes_leg.plot(x,x**2,label="X Squared")
axes_leg.plot(x,x**3,label="X Cubed")

# Displays info head based on label in axes.plot and loc represents location of that info head - value is std or can provide axis loc
axes_leg.legend(loc=9)
```

# Plot appearance

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

# can use hexa for color, linewidth as lw, linestyle as ls
axes.plot(x,y,color='purple',linewidth=3, alpha=0.7, linestyle='-', marker='o',markersize=20
         ,markerfacecolor='yellow', markeredgewidth=5,markeredgecolor='green')
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

# can use hexa for color, linewidth as lw, linestyle as ls
axes.plot(x,y,color='purple',linewidth=3, alpha=0.7, linestyle='--',marker='*')

# to limit x - zoom into x-axis
axes.set_xlim([0,5])
axes.set_ylim([0,6])
```
