```python
import seaborn as sas
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#Loading DataSet
tips = sas.load_dataset("tips")
```

```python
tips.head()
```

```python
# Distrubuted Graph - used on univariable - one variable
# kde - kernel Descity estimation - sum of normal distrubution
sas.distplot(tips['total_bill'],kde=False,bins=40)
```

```python
# Joint Graph - Used to compare any two variable
sas.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
# pairplot graph - Used to compare every variable in DataFrame
sas.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sas.rugplot(tips['total_bill'])
```

```python
sas.kdeplot(tips['total_bill'])
```

Categorical plotting - non-numeric and can be categerise

```python
import numpy as np
```

```python
sas.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) # display based on estimator default is mean(average)
```

```python
sas.countplot(x='sex',data=tips) # just count occurance
```

```python
sas.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sas.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sas.stripplot(x='day',y='total_bill',data=tips)
```

```python
sas.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# Factor Plot is general plot func. can plot any type of plot
sas.factorplot(x='day',y='total_bill',data=tips,kind='box')
```

```python
# Factor Plot is general plot func. can plot any kind of plot
sas.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sas.regplot(x='tip',y='total_bill',data=tips)
```

# matrix plot

```python
flights = sas.load_dataset('flights')
```

```python
flights.head()
```

```python
# In matrix plot, there should be a matrix/table like data 
sas.heatmap(data=tips.corr(),cmap='coolwarm')
```

```python
pivat_flight = flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
sas.heatmap(data=pivat_flight,cmap='magma',linecolor='white',linewidths=3)
```

```python
sas.clustermap(data=pivat_flight)
```

```python
sas.clustermap(data=pivat_flight, standard_scale=1) # standard_scale represents max_scale 
```

# Grid

```python
iris = sas.load_dataset('iris')
iris.head()
```

```python
# Grid which produces background grid/lines on sheet structure 

t = sas.PairGrid(iris)
t.map(plt.plot)
```

```python
t = sas.PairGrid(iris)
t.map_diag(plt.hist)
t.map_upper(plt.scatter)
t.map_lower(sas.kdeplot)
```

# Facet Grid is used to present categorical / featured Data

```python
g = sas.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.hist, 'total_bill')
```

# regression plot

```python
# here x = x-axis, y = y-axis but col represents no. of plots in a col and row represents no. of plots in a row
sas.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='smoker')
```
